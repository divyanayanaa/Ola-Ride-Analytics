```python
import pandas as pd
```

```python
path="OLA_DataSet.xlsx"
df=pd.read_excel(path,sheet_name="July")
```

```python
df=df.drop(columns=['Vehicle Images'])
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df['Payment_Method'] = df['Payment_Method'].fillna("Unknown")
```

```python
df['V_TAT'] = df['V_TAT'].fillna("NaN")
```

```python
df['C_TAT'] = df['C_TAT'].fillna("NaN")
```

```python
df['Canceled_Rides_by_Customer'] = df['Canceled_Rides_by_Customer'].fillna("NaN")
```

```python
df['Canceled_Rides_by_Driver'] = df['Canceled_Rides_by_Driver'].fillna("NaN")
```

```python
df['Incomplete_Rides'] = df['Incomplete_Rides'].fillna("NaN")
df['Incomplete_Rides_Reason'] = df['Incomplete_Rides_Reason'].fillna("NaN")
```

```python
df['Driver_Ratings'] = df['Driver_Ratings'].fillna("NaN")
df['Customer_Rating'] = df['Customer_Rating'].fillna("NaN")
```

```python
print(df.isnull().sum())
```

```python
df['Booking_Status'] = df['Booking_Status'].str.title()
df['Vehicle_Type'] = df['Vehicle_Type'].str.title()
```

```python
df['Date'] = pd.to_datetime(df['Date']).dt.date
```

```python
# Ride Outcome: Success / Cancelled / Incomplete
df['Ride_Outcome'] = df.apply(
    lambda x: "Incomplete" if str(x['Incomplete_Rides']).strip().lower() == "yes"
              else ("Cancelled" if "Cancel" in x['Booking_Status'] else "Success"),
    axis=1
)
```

```python
# Revenue per km (avoid divide-by-zero)
df['Revenue_per_km'] = df.apply(
    lambda x: x['Booking_Value']/x['Ride_Distance'] if x['Ride_Distance'] > 0 else None,
    axis=1
)
```

```python
# Binary flag for cancellations
df['Is_Cancelled'] = df['Booking_Status'].str.contains("Cancel").astype(int)
```

```python
print("Unique Booking Statuses:", df['Booking_Status'].unique())
print("Unique Payment Methods:", df['Payment_Method'].unique())
print("Unique Vehicle Types:", df['Vehicle_Type'].unique())
print("Duplicate Booking IDs:", df['Booking_ID'].duplicated().sum())
```
